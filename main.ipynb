```python
df = pd.read_excel('low_data.xlsx', sheet_name="Sheet1")
durasi = pd.DataFrame(df["Durasi"][:100])
durasi.plot()
plt.title("Durasi Untuk Seratus Data Pertama", pad=15, fontsize=14, color='black')
plt.xlabel("Data Panggilan")
plt.ylabel("Durasi Panggilan")
plt.show()
```

```python
df = pd.read_excel('low_data.xlsx', sheet_name="Sheet1")
jakarta_tel = df[df["Area tujuan"] == "Jakarta, (TEL)"]
jakarta_tel.head()
```

```python
jakarta_sorted = jakarta_tel.sort_values(by=['Waktu pemanggilan'])
jakarta_sorted
```

```python
jakarta_sorted.plot(figsize=(20,4), kind="bar", x="Waktu pemanggilan", y="Durasi", color="blue")

plt.show()
```
